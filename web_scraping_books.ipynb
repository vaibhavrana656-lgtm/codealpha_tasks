# CodeAlpha Task 1 — Web Scraping

## Objective
Create a custom dataset from **Books to Scrape**, a public website designed for web-scraping practice.

The project collects:
- title
- price
- star rating
- stock status
- product URL
- image URL

The complete live scraper is available in `scraper.py`. The bundled CSV allows offline review and analysis.

```python
from pathlib import Path
import pandas as pd

DATA_PATH = Path("data/books_scraped.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing values:\n", df.isna().sum())
print("\nDuplicate product URLs:", df["product_url"].duplicated().sum())
```

```python
summary = {
    "books_collected": len(df),
    "average_price_gbp": round(df["price_gbp"].mean(), 2),
    "median_price_gbp": round(df["price_gbp"].median(), 2),
    "average_rating": round(df["rating_numeric"].mean(), 2),
    "in_stock_percentage": round(df["in_stock"].mean() * 100, 2),
}
summary
```

## Completed result
- Records collected: **993**
- Average price: **£35.13**
- Median price: **£36.17**
- Average rating: **3.13/5**
- Duplicate product URLs: **0**
