# CodeAlpha Task 2 — Exploratory Data Analysis

## Questions
1. What is the typical price range?
2. How are books distributed across ratings?
3. Do prices differ across star-rating groups?
4. Is title length associated with price?
5. Are there missing values, duplicates, or outliers?

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kruskal, spearmanr

df = pd.read_csv(Path("../task_1_web_scraping/data/books_scraped.csv"))
df.head()
```

```python
quality = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "missing": df.isna().sum(),
    "unique": df.nunique()
})
quality
```

```python
df[["price_gbp", "rating_numeric", "title_length"]].describe().round(2)
```

```python
df.groupby("rating_numeric")["price_gbp"].agg(
    ["count", "mean", "median", "min", "max"]
).round(2)
```

```python
groups = [
    group["price_gbp"].dropna()
    for _, group in df.groupby("rating_numeric")
]
kw_stat, kw_p = kruskal(*groups)
rho, rho_p = spearmanr(df["price_gbp"], df["title_length"])

print(f"Kruskal-Wallis p-value: {kw_p:.6f}")
print(f"Spearman correlation: {rho:.4f}")
print(f"Spearman p-value: {rho_p:.6f}")
```

## Findings
- Rows analyzed: **993**
- Mean price: **£35.13**
- Median price: **£36.17**
- Kruskal-Wallis p-value: **0.530056**
- Price/title-length Spearman correlation: **0.0234**
- Correlation p-value: **0.461823**

The results do not show a statistically significant difference in price across ratings, and title length has almost no relationship with price.
