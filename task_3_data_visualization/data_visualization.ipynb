# CodeAlpha Task 3 — Data Visualization

## Goal
Turn raw book data into visual formats that communicate price, rating, stock, and product-level insights clearly.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(Path("../task_1_web_scraping/data/books_scraped.csv"))
{
    "total_books": len(df),
    "average_price": round(df["price_gbp"].mean(), 2),
    "median_price": round(df["price_gbp"].median(), 2),
    "average_rating": round(df["rating_numeric"].mean(), 2),
}
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["price_gbp"], bins=20, edgecolor="black")
plt.title("Distribution of Book Prices")
plt.xlabel("Price (GBP)")
plt.ylabel("Number of Books")
plt.show()
```

```python
counts = df["rating_numeric"].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(counts.index.astype(str), counts.values)
plt.title("Books by Star Rating")
plt.xlabel("Star Rating")
plt.ylabel("Number of Books")
plt.show()
```

```python
avg_price = df.groupby("rating_numeric")["price_gbp"].mean()
plt.figure(figsize=(8, 5))
plt.bar(avg_price.index.astype(str), avg_price.values)
plt.title("Average Price by Star Rating")
plt.xlabel("Star Rating")
plt.ylabel("Average Price (GBP)")
plt.show()
```
